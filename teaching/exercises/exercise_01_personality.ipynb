```python
# ============================================================
# ร้านค้าและโปรโมชั่น (ไม่ต้องแก้)
# ============================================================

STORES = [
    {"name": "CJ More", "type": "supermarket", "distance": "ใกล้", "price": "ปานกลาง",
     "promos": ["ลดผงซักฟอก 15%"], "quality": "ปานกลาง"},
    {"name": "Big C", "type": "hypermarket", "distance": "ไกล", "price": "ถูก",
     "promos": ["ลดนม 30%", "ลดไข่ 10%"], "quality": "ดี"},
    {"name": "7-Eleven", "type": "convenience", "distance": "ใกล้มาก", "price": "แพง",
     "promos": ["กาแฟลด 50%"], "quality": "ปานกลาง"},
    {"name": "Tops", "type": "supermarket", "distance": "ปานกลาง", "price": "แพง",
     "promos": ["สินค้า premium ลด 20%"], "quality": "ดีมาก"},
]


def score_store(personality: dict, store: dict) -> float:
    """
    คำนวณ score ว่า agent ชอบร้านนี้แค่ไหน

    สูตร (simplified):
    - ถ้า price_sensitivity สูง -> ชอบร้านราคาถูก
    - ถ้า convenience_preference สูง -> ชอบร้านใกล้
    - ถ้า promotion_attraction สูง -> ชอบร้านมีโปรเยอะ
    - ถ้า quality_preference สูง -> ชอบร้านคุณภาพดี
    """
    score = 0.0
    p = personality

    # Price factor
    price_scores = {"ถูก": 1.0, "ปานกลาง": 0.5, "แพง": 0.2}
    score += p["price_sensitivity"] * price_scores.get(store["price"], 0.5)

    # Convenience factor
    dist_scores = {"ใกล้มาก": 1.0, "ใกล้": 0.7, "ปานกลาง": 0.4, "ไกล": 0.2}
    score += p["convenience_preference"] * dist_scores.get(store["distance"], 0.5)

    # Promotion factor
    score += p["promotion_attraction"] * (len(store["promos"]) * 0.3)

    # Quality factor
    quality_scores = {"ดีมาก": 1.0, "ดี": 0.7, "ปานกลาง": 0.4}
    score += p["quality_preference"] * quality_scores.get(store["quality"], 0.5)

    return score


print(f"ร้านค้า {len(STORES)} ร้านพร้อมแล้ว!")
for s in STORES:
    print(f"  - {s['name']} ({s['type']}) | ราคา: {s['price']} | โปร: {', '.join(s['promos'])}")
```

## Run Experiment!

Run cell ด้านล่างเพื่อเปรียบเทียบ 3 agents

ดูว่า agent ที่มี personality ต่างกัน **เลือกร้านต่างกันอย่างไร**

```python
# ============================================================
# Run Experiment: เปรียบเทียบ 3 agents
# ============================================================

TRAIT_LABELS = {
    "price_sensitivity": "ดูราคา",
    "brand_loyalty": "ภักดีแบรนด์",
    "quality_preference": "ชอบของดี",
    "convenience_preference": "สะดวกสำคัญ",
    "promotion_attraction": "ชอบโปร",
    "impulse_buying": "ซื้อตามใจ",
}

agents = [AGENT_ORIGINAL, AGENT_RICH, AGENT_IMPULSE]

# เช็คว่าแก้ TODO หรือยัง
if sum(AGENT_RICH["personality"].values()) < 1.0:
    print("WARNING: ยังไม่ได้แก้ TODO!")
    print("  กรุณาใส่ค่า personality ใน AGENT_RICH และ AGENT_IMPULSE")
    print("  แล้ว Run cell นี้ใหม่")
    print()

print("=" * 70)
print("  PERSONALITY EXPERIMENT: เปลี่ยน personality -> เปลี่ยนร้านที่เลือก")
print("=" * 70)

for agent in agents:
    print(f"\n{'─'*70}")
    print(f"  {agent['name']}")
    print(f"{'─'*70}")

    # แสดง personality
    p = agent["personality"]
    print("  Personality:")
    for trait, val in p.items():
        bar = "█" * int(val * 10) + "░" * (10 - int(val * 10))
        label = TRAIT_LABELS.get(trait, trait)
        print(f"    {label:<15} {bar} {val:.1f}")

    # คำนวณ score สำหรับแต่ละร้าน
    print(f"\n  Store Scores:")
    store_scores = []
    for store in STORES:
        score = score_store(p, store)
        store_scores.append((store["name"], score))

    store_scores.sort(key=lambda x: x[1], reverse=True)

    for i, (name, score) in enumerate(store_scores):
        bar = "█" * int(score * 15)
        marker = " <-- CHOSEN!" if i == 0 else ""
        print(f"    {name:<15} {score:.2f}  {bar}{marker}")

    winner = store_scores[0]
    print(f"\n  -> ตัดสินใจ: ไป {winner[0]} (score = {winner[1]:.2f})")
```
