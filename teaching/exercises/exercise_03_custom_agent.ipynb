## TODO: สร้าง Memories ให้ Agent (อย่างน้อย 5 อัน)

แต่ละ memory ประกอบด้วย:
- **เนื้อหา** (str): เหตุการณ์ที่เกิดขึ้น
- **importance** (1-10): ความสำคัญ
- **ชั่วโมงที่แล้ว** (int): เกิดขึ้นกี่ชั่วโมงที่แล้ว

````python
# ============================================================
# TODO: สร้าง Memories ให้ agent ของคุณ (อย่างน้อย 5 อัน)
# ============================================================
#
# Format: ("เนื้อหา", importance, ชั่วโมงที่แล้ว)

MY_MEMORIES = [
    # TODO: ใส่ memories ตรงนี้
    # ("ไป Big C เจอโปรลดนม 30% ซื้อมาเยอะ", 7, 72),
    # ("ร้าน 7-11 ใต้คอนโดเปิดใหม่ สะดวกมาก", 6, 48),
    # ("ลืมซื้อไข่! ต้องกลับไปซื้อใหม่", 8, 24),
    # ("เห็นใบปลิว CJ More ลดครบรอบ", 5, 12),
    # ("ตู้เย็นว่าง ต้องไปซื้อของแล้ว", 9, 2),
]

print(f"สร้าง {len(MY_MEMORIES)} memories แล้ว")
if len(MY_MEMORIES) < 5:
    print("WARNING: ควรมีอย่างน้อย 5 memories!")
````

````python
# ============================================================
# Simulation Code (ไม่ต้องแก้)
# ============================================================

from datetime import datetime, timedelta
from dataclasses import dataclass
from typing import Optional


@dataclass
class Memory:
    content: str
    created_at: datetime
    importance: int
    memory_type: str = "observation"


class MemoryStream:
    RECENCY_DECAY = 0.99
    def __init__(self, name):
        self.name = name
        self.memories = []

    def add(self, content, importance, created_at=None):
        mem = Memory(content, created_at or datetime.now(), importance)
        self.memories.append(mem)

    def retrieve(self, query, top_k=5, current_time=None):
        ct = current_time or datetime.now()
        results = []
        for m in self.memories:
            hrs = (ct - m.created_at).total_seconds() / 3600
            rec = self.RECENCY_DECAY ** hrs
            imp = m.importance / 10.0
            qw = set(query.lower().split())
            mw = set(m.content.lower().split())
            rel = min(1.0, len(qw & mw) / len(qw)) if qw else 0
            results.append((m, rec * imp * rel))
        results.sort(key=lambda x: x[1], reverse=True)
        return results[:top_k]


STORES = [
    {"name": "CJ More", "type": "supermarket", "distance": "ใกล้", "price": "ปานกลาง",
     "promos": ["ลดผงซักฟอก 15%"]},
    {"name": "Big C", "type": "hypermarket", "distance": "ไกล", "price": "ถูก",
     "promos": ["ลดนม 30%", "ลดไข่ 10%"]},
    {"name": "7-Eleven", "type": "convenience", "distance": "ใกล้มาก", "price": "แพง",
     "promos": ["กาแฟลด 50%"]},
    {"name": "Tops", "type": "supermarket", "distance": "ปานกลาง", "price": "แพง",
     "promos": ["สินค้า premium ลด 20%"]},
    {"name": "Makro", "type": "wholesale", "distance": "ไกลมาก", "price": "ถูกมาก",
     "promos": ["ซื้อ 3 แถม 1"]},
]


def score_store(personality, store):
    score = 0.0
    price_scores = {"ถูกมาก": 1.0, "ถูก": 0.8, "ปานกลาง": 0.5, "แพง": 0.2}
    dist_scores = {"ใกล้มาก": 1.0, "ใกล้": 0.7, "ปานกลาง": 0.4, "ไกล": 0.2, "ไกลมาก": 0.1}
    score += personality["price_sensitivity"] * price_scores.get(store["price"], 0.5)
    score += personality["convenience_preference"] * dist_scores.get(store["distance"], 0.5)
    score += personality["promotion_attraction"] * (len(store["promos"]) * 0.3)
    return score


print("Simulation code loaded!")
````

## Run Simulation!

Run cell ด้านล่างเพื่อดูว่า Agent ของคุณตัดสินใจอย่างไร

````python
# ============================================================
# Run Simulation
# ============================================================

agent = MY_AGENT
NOW = datetime(2024, 1, 13, 10, 0)

# Validate
if not agent["name"]:
    print("X ยังไม่ได้กรอกข้อมูล agent! กรุณากรอก MY_AGENT ก่อน")
elif not MY_MEMORIES:
    print("X ยังไม่ได้สร้าง memories! กรุณากรอก MY_MEMORIES ก่อน")
else:
    print("=" * 60)
    print(f"  YOUR AGENT: {agent['name']}")
    print("=" * 60)

    # Profile
    print(f"\n  {agent['name']} ({agent['job']}, อายุ {agent['age']})")
    print(f"  ครอบครัว: {agent['household_size']} คน")
    print(f"  รายได้: {agent['monthly_income']:,} บาท/เดือน")
    print(f"  เป้าหมาย: {agent['financial_goal']}")
    if agent['traits']:
        print(f"  นิสัย: {', '.join(agent['traits'])}")
    if agent['pain_points']:
        print(f"  ไม่ชอบ: {', '.join(agent['pain_points'])}")
    print(f"  Background: {agent['background']}")

    # Personality
    print(f"\n  Personality:")
    trait_labels = {
        "price_sensitivity": "ดูราคา",
        "brand_loyalty": "ภักดีแบรนด์",
        "quality_preference": "ชอบของดี",
        "convenience_preference": "สะดวกสำคัญ",
        "promotion_attraction": "ชอบโปร",
        "impulse_buying": "ซื้อตามใจ",
    }
    for trait, val in agent["personality"].items():
        bar = "█" * int(val * 10) + "░" * (10 - int(val * 10))
        label = trait_labels.get(trait, trait)
        print(f"    {label:<15} {bar} {val:.1f}")

    # Memory
    print(f"\n  Memories ({len(MY_MEMORIES)} อัน):")
    ms = MemoryStream(agent["name"])
    for content, imp, hours_ago in MY_MEMORIES:
        ms.add(content, imp, created_at=NOW - timedelta(hours=hours_ago))
        print(f"    - [{imp}/10] {content}")

    # Retrieve
    query = "ซื้อ ของ ร้าน โปรโมชั่น"
    retrieved = ms.retrieve(query, top_k=3, current_time=NOW)
    print(f"\n  Retrieved Memories (query: \"{query}\"):")
    for mem, score in retrieved:
        print(f"    [{score:.4f}] {mem.content}")

    # Store Selection
    print(f"\n  Store Scores:")
    store_scores = []
    for store in STORES:
        score = score_store(agent["personality"], store)
        store_scores.append((store["name"], score, store))
    store_scores.sort(key=lambda x: x[1], reverse=True)

    for i, (name, score, store) in enumerate(store_scores):
        bar = "█" * int(score * 15)
        marker = " <-- CHOSEN!" if i == 0 else ""
        print(f"    {name:<15} {score:.2f}  {bar}{marker}")

    winner = store_scores[0]
    print(f"\n  {agent['name']} ตัดสินใจไป {winner[0]}!")
    print(f"\n{'='*60}")
````
